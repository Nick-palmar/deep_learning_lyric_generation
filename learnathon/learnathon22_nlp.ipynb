```python
import pandas as pd
```

```python
lyrics_df = pd.read_csv('sample_data/lyrics-data.csv', error_bad_lines=False, engine="python")
lyrics_df.head()
```

```python
artist_df = pd.read_csv('sample_data/artists-data.csv')
artist_df.head()
```

```python
lyrics_df.rename(columns={'ALink': 'Link'}, inplace=True)
```

```python
lyrics_df.head()
```

```python
combined_df = lyrics_df.merge(artist_df, on='Link')[['Lyric', 'Artist', 'language', 'Songs', 'SName']]
combined_df.head()
```

```python
combined_df.drop_duplicates(subset=['SName'], inplace=True)
```

```python
english_df = combined_df.loc[combined_df['language'] == 'en']
english_df.head()
```

```python
pd.Series(pd.Series(english_df.loc[english_df['Songs'] > 250, 'Artist']).unique()).sort_values()
```

```python
elvis_df = combined_df.loc[combined_df['Artist'] == 'Elvis Presley']
elvis_df.head()
```

```python
lang_model_block = DataBlock(
    blocks=TextBlock.from_df('elvis_df', seq_len=72, is_lm=True),
    get_items=ColReader('Lyric')
)
lang_model_block
```

```python
dls_lm = lang_model_block.dataloaders(elvis_df, bs=128)
dls_lm.show_batch(max_n=2)
```

```python
learn = language_model_learner(dls_lm, AWD_LSTM, drop_mult=0.3, metrics=accuracy)
learn.fit_one_cycle(5, 0.004)
learn.unfreeze()
# train for 20 more epochs on the new lr
learn.fit_one_cycle(20, lr_max=slice(3e-6, 3e-4))
```

```python
def get_most_complex(start_text, preds):
  max_len = 0
  max_i = -1
  for i, pred in enumerate(preds):
    pred_cardinality = len(set(pred.split()))
    if pred_cardinality > max_len:
      max_len = pred_cardinality
      max_i = i
  
  return_str = preds[max_i]

  val = -1
  occurrence = len(start_text.split())
  for i in range(0, occurrence):
    val = return_str.find(' ', val + 1)

  return start_text + return_str[val:return_str.rfind('.')+1]
```

```python
start_text = "Fun stuff song"
words = 60
sentences = 5
preds = [learn.predict(start_text, words, temperature=0.75)
         for sentence in range(sentences)]

get_most_complex(start_text, preds)
```
